#### 2.2 Creation of validation set

```python
# Get the list of all subdirectories in the train dataset directory
subdirectories = os.listdir(train_path)

# Create the validation directory if it doesn't exist
if not os.path.exists(validation_path):
    os.makedirs(validation_path)

# Iterate through each subdirectory
for subdirectory in subdirectories:
    # Create the full path to the subdirectory in the train dataset
    subdirectory_path_train = os.path.join(train_path, subdirectory)
    
    # Create the full path to the subdirectory in the validation dataset
    subdirectory_path_validation = os.path.join(validation_path, subdirectory)
    
    # Check if the item in the train folder is a directory
    if os.path.isdir(subdirectory_path_train):
        # Get the list of image file names in the subdirectory
        image_files = os.listdir(subdirectory_path_train)
        
        # Create the subdirectory in the validation dataset
        if not os.path.exists(subdirectory_path_validation):
            os.makedirs(subdirectory_path_validation)
        
        # Move a portion of the images to the validation dataset
        num_images_validation = int(len(image_files) * 0.2)  # Adjust the split ratio as needed
        images_to_move = image_files[:num_images_validation]
        for image_file in images_to_move:
            source_path = os.path.join(subdirectory_path_train, image_file)
            destination_path = os.path.join(subdirectory_path_validation, image_file)
            shutil.move(source_path, destination_path)
```

#### 2.5 Load Image Datasets and Apply Augmentations

```python
image_width = 150
image_height = 150
BBatch_size = 32
```

```python
datagen=ImageDataGenerator(rescale=1/255.,
                           zoom_range=0.2,
                           horizontal_flip=True,
                           validation_split=0.15)

train_generator=datagen.flow_from_directory(train_path,
                                       target_size=(image_width,image_height),
                                       batch_size=BBatch_size,
                                       shuffle=True,
                                       class_mode='categorical')
test_generator=datagen.flow_from_directory(test_path,
                                      shuffle=False,
                                      batch_size=BBatch_size,
                                      target_size=(image_width,image_height),
                                      class_mode="categorical")

validation_generator=datagen.flow_from_directory(validation_path,
                                    shuffle=False,
                                    batch_size=BBatch_size,
                                    target_size=(image_width,image_height),
                                    class_mode='categorical')
```

```python
validation_data_generator = ImageDataGenerator(rescale=1.0/255)
validation_generator = validation_data_generator.flow_from_directory(validation_path,
                                                   target_size=(image_width, image_height),
                                                   batch_size=BBatch_size,
                                                   class_mode='categorical',
                                                   shuffle=True)
```

```python
test_data_generator = ImageDataGenerator(rescale=1.0/255)
test_generator = test_data_generator.flow_from_directory(test_path,
                                                   target_size=(image_width, image_height),
                                                   batch_size=BBatch_size,
                                                   class_mode='categorical',
                                                   shuffle=True)
```

### 4. Training a CNN Model

```python
model=Sequential()

model.add(Conv2D(32, (3,3), activation='relu', input_shape = (image_height,image_width,3), padding="same"))
model.add(MaxPooling2D())

model.add(Conv2D(64, (3,3), activation='relu', padding="same"))
model.add(MaxPooling2D())

model.add(Conv2D(128,(3,3), activation='relu', padding="same"))
model.add(MaxPooling2D())

model.add(Conv2D(256,(3,3), activation='relu', padding="same", strides=2))
model.add(MaxPooling2D())

model.add(Flatten())

model.add(Dense(256, activation="relu"))
# model.add(Dropout(.3))
model.add(Dense(6, activation="softmax"))
```

```python
model.compile(optimizer="adam",loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
print(model.summary())
```

```python
cc_model = model.fit(train_generator, epochs=15, validation_data=validation_generator,
                       verbose=1)
```

### 5. Plotting the Model Metrics
